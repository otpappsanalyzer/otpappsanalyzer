```python
names=['app_id','url','class','language','message']
ml_result = pd.read_csv(pp_result_against_ml,header=None,names=names)
ml_result
```

```python
clean_from_uncrawled = ml_result[ml_result["class"].str.contains("None")==False]
clean_from_uncrawled
```

```python
clean_non_english = clean_from_uncrawled[clean_from_uncrawled["language"].str.contains("en")==True]
clean_non_english
```

```python
def pct(x):
    N = 182 # number of OTP apps
    res = (x/N)*100
    return res
```

```python
pp_positive = clean_non_english[clean_non_english["class"].str.contains("True")==True]
print("Number of + Apps :" + str(len(pp_positive))+ " | Percentage :"+ str(pct(len(pp_positive))))
# Number of OTP apps POSITIVE mentioning that they share the user information to the third party or
# They mentioned to share the user information to the third party
```

```python
pp_negative = clean_non_english[clean_non_english["class"].str.contains("False")==True]
print ("Number of apps "+ str(len(pp_negative)) + " | Percentage : " + str(pct(len(pp_negative))))
# Number of OTP apps NEGATIVE mentioning that they share the user information to the third party or
# They mentioned NOT to share the user information to the third party
```
