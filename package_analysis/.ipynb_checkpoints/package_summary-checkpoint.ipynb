```python
import pandas as pd
path = '/home/budi/OTP_project/OTP_code/package_analysis/noncompiler_list.csv'

apkid_df = pd.read_csv(path)
# apkid_df
apkid_df['app_id'] = apkid_df['file_name'].str.rstrip('apk')
apkid_df['app_id'] = apkid_df['app_id'].str.rstrip('.')
pd.set_option('display.max_rows', None)
apkid_df
```

```python
# This script is use to create anti vm df
anti_vm_df = apkid_df[apkid_df['classes']=='anti_vm']
anti_vm_df = anti_vm_df[['app_id','classes','component']]

detail_anti_vm_method = []
for x, line in anti_vm_df.iterrows():
    components = line['component'].strip('][').split(', ')
    for item in components:
        item_detail = {'app_id':line['app_id'],'method':item}
        if item_detail not in detail_anti_vm_method:
            detail_anti_vm_method.append(item_detail)

detail_anti_vm_df = pd.DataFrame(detail_anti_vm_method)
detail_anti_vm_df
```

```python
group_by_anti_vm_method = detail_anti_vm_df.groupby(['method']).size().reset_index(name='count').sort_values(by=['count'],ascending=False)
group_by_anti_vm_method
```
